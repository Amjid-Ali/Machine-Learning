```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("./Social_Network_Ads.csv")
# dataset
```

```python
features=dataset.iloc[:,[2,3]].values
label=dataset.Purchased
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,label, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(C=100,kernel = 'linear', random_state = 0)
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
print(model.score(X_train,y_train))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
confidence=accuracy_score(y_test,y_pred)
```

```python
confidence
```

```python
cm=confusion_matrix(y_test,y_pred)
```

# kernal svm

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf',gamma=6,C=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```
